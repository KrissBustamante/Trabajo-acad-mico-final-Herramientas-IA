
# **Generacion la base de datos**
Aqui se muestra el siguiente notebook de como se creo la base de datos


```python
import sqlite3

# Crear base de datos en memoria o archivo temporal
conn = sqlite3.connect("gimnasio.db")
cursor = conn.cursor()

# Crear tabla y cargar datos
cursor.executescript("""
DROP TABLE IF EXISTS ejercicios;
CREATE TABLE ejercicios (
    id INTEGER PRIMARY KEY,
    nombre TEXT,
    calorias_quemadas INT
);

INSERT INTO ejercicios (nombre, calorias_quemadas) VALUES
('Caminata', 200),
('Trote', 450),
('Ciclismo', 300),
('Pesas', 250),
('Yoga', 180);
""")

conn.commit()

```

```python
import pandas as pd

# Leer datos con Pandas
df_sql = pd.read_sql("SELECT * FROM ejercicios", conn)

# Mostrar
df_sql
```

```python
df_sql.to_csv("dataset_sql.csv", index=False)
```
